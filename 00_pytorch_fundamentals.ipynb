## Matrix Multiplication

- The no of columns of the `first matrix` must be the same as the no of rows of the `second matrix`

- The resulting matrix will have a shape of (`no of rows of the first matrix`, `no of columns of the second matrix`)

- `tensor1 @ tensor2` or `torch.matmul(tensor1, tensor2)`

```python
tensor = torch.tensor([1,2,3])
tensor.shape
```

```python
# Element-wise operation
tensor * tensor
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor), tensor @ tensor
```

```python
%%time
# Matric Multiplication by hand
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]

value
```

```python
%%time
torch.matmul(tensor, tensor)
```

## Shape Errors

```python
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]], dtype = torch.float32)

tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]], dtype = torch.float32)

torch.matmul(tensor_A, tensor_B)
```

```python
print(tensor_A)
print(tensor_B)
```

```python
print(tensor_A)
print(tensor_B.T)
print(torch.transpose(tensor_B, 0, 1))
```

```python
print(f'Original Shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n')
print(f'New Shapes: tensor_A: {tensor_A.shape}, tensor_B.T = {tensor_B.T.shape}\n')
print(f'Multipling: {tensor_A.shape} * {tensor_B.shape}\n')
print('Output:\n')
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f'\nOutput Shape: {output.shape}')
```

We can also use `torch.mm()` instead of `torch.matmul()`

```python
torch.mm(tensor_A, tensor_B.T)
```

## Reshaping, staking, squeezing and unsqueezing

```python
import torch
x = torch.arange(1., 8.)
x, x.shape
```

```python
x_reshaped = x.reshape(shape = (1,7))
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1,7)
z, z.shape
```

```python
z[:, 0] = 5
z,x
```

```python
x_stacked = torch.stack([x,x,x,x,x], dim = 0)
x_stacked
```

```python
print(f'Previous tensor: {x_reshaped}')
print(f'Previous shape: {x_reshaped.shape}')

x_squeezed = x_reshaped.squeeze()
print(f'\nNew tensor: {x_squeezed}')
print(f'New shape: {x_squeezed.shape}')
```

```python
print(f'Previous tensor: {x_squeezed}')
print(f'Previous shape: {x_squeezed.shape}')

x_unsqueezed = x_squeezed.unsqueeze(dim = 0)
print(f'\nNew Tensor: {x_unsqueezed}')
print(f'New shape: {x_unsqueezed.shape}')
```

```python
x_original = torch.rand(size = (224,224, 3))

x_permuted = x_original.permute(2,0,1)

print(f'Previous shape: {x_original.shape}')
print(f'New shape: {x_permuted.shape}')
```

# Reproducibility
> Trying to take random out of random

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(f'Tensor_A: {random_tensor_A}\n')
print(f'Tensor_B: {random_tensor_B}\n')
print(f'is Tensor_A equal to Tensor_B: \n {random_tensor_A == random_tensor_B}')
```

> Let's make them the same

```python
random_seed = 42
torch.manual_seed(seed = random_seed)
random_tensor_C =  torch.rand(3,4)

torch.random.manual_seed(seed = random_seed)
# torch.manual_seed(seed = random_seed) also works

random_tensor_D = torch.rand(3,4)

print(f'Tensor_C: {random_tensor_C}\n')
print(f'Tensor_D: {random_tensor_D}\n')
print(f'is Tensor_C equal to Tensor_D: \n {random_tensor_C == random_tensor_D}')
```
